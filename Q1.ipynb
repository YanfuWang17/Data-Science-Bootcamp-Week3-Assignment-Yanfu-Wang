```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD')
```

```python
print(df.head())
print(df.shape)
```

```python
df["hour_beginning"] = pd.to_datetime(df['hour_beginning'])
df['day_name'] = df['hour_beginning'].dt.day_name()
print(df.head())
```

```python
# I am not quite sure which graph is the line graph, so I plot more than one graphs.
plt.figure(figsize=(7, 3))
sns.histplot(data=df, x='day_name')
plt.title('Histogram of Pedestrians in Weekdays')
plt.xlabel('Weekdays')
plt.ylabel('Pedestrian Count')
plt.show()
```

```python
day_counts = df.groupby(df['day_name'])['Pedestrians'].sum()
plt.figure(figsize=(12, 6))
day_counts.plot(kind='bar', color='orange')
plt.title('Total Pedestrian Counts In Weekdays')
plt.xlabel('Days')
plt.ylabel('Pedestrian Count')
plt.grid(axis='y')  #grid created alone y axis
plt.tight_layout()
plt.show()
```
